```python
def outliers(dt, ft):
    Q1 = dt[ft].quantile(0.25)
    Q3 = dt[ft].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    ls = dt.index[(dt[ft] < lower_bound) | (dt[ft] > upper_bound)]
    return ls
```

```python
index_list = []
for feature in ['housing_median_age', 'households', 'median_income', 'median_house_value', 'rooms_per_household', 'bedrooms_per_room', 'population_per_household']:
    index_list.extend(outliers(data, feature))
```

```python
def remove(dt, ls):
    ls = sorted(set(ls))
    dt = dt.drop(ls)
    return dt
```

```python
data_ = remove(data, index_list)
```

# Modelling

```python
x = data_.loc[:,['housing_median_age', 'households', 'median_income', 'rooms_per_household','bedrooms_per_room', 'population_per_household']]
y = data_.median_house_value
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state = 15)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
print("Intercept is", lr.intercept_, "and Coefficients are", lr.coef_)
```

```python
pred = lr.predict(x_test)
```

# Model Evaluation

```python
from sklearn.metrics import r2_score
r_squared = r2_score(y_test,pred)

adjusted_r_squared = 1 - (1-r_squared)*(len(y) -1)/(len(y)-x.shape[1]-1)
                                        
print('Adjsuted R Squared:%', round(adjusted_r_squared*100,2), 'R Sqaured: %', round(r_squared*100,2))
```
